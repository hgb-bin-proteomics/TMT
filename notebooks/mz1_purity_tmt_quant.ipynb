```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("mz1_purity_tmt_quant.parquet")
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.hist(df["Co-Isolation Purity"], bins=50, density=False, histtype="bar", fill=True)
ax.set_ylabel("Number of identifications")
ax.set_xlabel("Purity")
```
